```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov13n.pt") 
```

#Train_the_Model

```python
results = model.train(
    data="coco8.yaml",   
    epochs=3,
    imgsz=512,
    batch=4,
    device="cpu"
)
```

```python
metrics = model.val(data="coco8.yaml")
```

```python
results = model("Object_Detection.png")
```

```python
results[0].show()
```

#Notes -
#1. (model=) argument in Ultralytics accepts either a .pt pretrained weight file or a .yaml architecture file path/name.
#2. (data=) argument takes a path or the name of a packaged dataset YAML such as coco.yaml or the small sample coco8.yaml.
#3. The official COCO dataset definition (coco8.yaml) is maintained in the Ultralytics repo; pointing to data=coco.yaml.
#4. So yolov13 model Not in the official Ultralytics package (yet). They removed it for some errors, Ultralytics‚Äô current public builds only
#auto‚Äëdownload pretrained weights up to YOLOv12‚ÄîYOLOv13 is not on that list, so a plain call like YOLO("yolov13n.pt") will fail to resolve 
#a model from their CDN.
#5. So, iMoonLab provides the required pre-trained model for YOLOv13, I use that here.
